```python
import pandas as pd 
```

```python
df=pd.read_csv("Titanic-Dataset.csv")
```

```python
# A.
df1=df[["Name","Sex","Age","Fare","Survived"]]

# B.
df2=df[(df["Sex"]== "female") & (df["Fare"]>30)][["Name","Sex","Age","Fare","Survived"]]

# C.
result = df.groupby("Pclass").agg({
    "Survived": "mean",
    "Fare": "mean",
    "Age": "mean"
})

result = result.rename(columns={      # Rename columns
    "Survived": "Survival Rate",
    "Fare": "Average Fare",
    "Age": "Average Age"
})
print(result)
```
