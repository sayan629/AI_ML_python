## Histograms

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Histograms
np.random.seed(0) #Normally, random numbers are different in every time 
                  #But when you use seed(0)you get the same random numbers every time


scores=np.random.normal(70,10,100) #follow ->Normal Distributions

bins=[0,30,70,90,100]
plt.hist(
         scores,
         color="skyblue",
         edgecolor="black",
         bins=bins
    )

plt.ylabel("# of Students")
plt.xlabel("scores")
# print(scores)
```

```python
marks = [
    25, 38, 42, 55, 61, 67, 72, 78, 83, 88,
    91, 95, 47, 53, 59, 64, 69, 74, 79, 85,
    92, 97, 33, 45, 58, 62, 71, 76, 81, 89
]
plt.hist(
    marks,
    color="cyan",
    edgecolor="black",
    bins=bins
)
plt.title("Distribution of marks for Students")
plt.ylabel("Number of Student")
plt.xlabel("Marks")
```

### Multiple Datasets in Histograms

```python
# Legit & Loud Transactions Distribution:
legit_transactions = [
    2.99, 5.49, 8.99, 12.50, 14.99, 19.99, 23.45, 29.99, 34.99, 39.50,
    45.00, 49.99, 55.25, 60.00, 75.99, 89.99, 120.50, 150.00, 199.99,
    249.99, 300.75, 450.00, 600.00, 850.00, 1200.00
]
fraud_transactions = [
    50, 100, 150, 200, 300, 500, 500, 750, 1000, 1000,
    1200, 1500, 1500, 2000, 2500, 3000, 3000
]

plt.hist(
    fraud_transactions,
    color="red",
    bins=20,
    label="Fraud",
    edgecolor="black"
)
plt.hist(
    legit_transactions,
    color="green",
    bins=20,
    label="Legit",
    edgecolor="black",
    alpha=0.5 
)
plt.title("Transaction Amount Distribution - Legit vs Fraud")
plt.xlabel("Amount (in $)")
plt.ylabel("Frequency")
plt.grid(True)
plt.legend()
```

### Histograms: axvline
- `axvline` is used to draw a **vertical reference line** on a plot.  
- It is commonly used to mark cutoffs, averages, pass marks, or thresholds.

```python
marks = [
    25, 38, 42, 55, 61, 67, 72, 78, 83, 88,
    91, 95, 47, 53, 59, 64, 69, 74, 79, 85,
    92, 97, 33, 45, 58, 62, 71, 76, 81, 89
]

plt.hist(
    marks,
    color="skyblue",
    edgecolor="black",
    bins=bins
)
plt.axvline(
    x=33,
    label="Passing Marks (33)",
    color="red",
    linewidth=2,
    linestyle="--"
)


plt.title("Distribution of Marks for Students.")
plt.ylabel("Number of Students")
plt.xlabel("Marks of Students")
plt.grid(True)
```

### Box Plots

- A Box Plot is a statistical visualization that summarizes the distribution of a dataset using five key numbers:

  - Minimum (lower non-outlier)  
  - First Quartile (Q1) – 25th Percentile  
  - Median (Q2) – 50th Percentile  
  - Third Quartile (Q3) – 75th Percentile  

  <br>
- Box Plot helps to detect outliers.  
- Helps to check spread (dispersion) of data.  
- Helps to identify skewness (data symmetry).

```python
data = [7, 8, 5, 6, 9, 4, 10, 12, 15,25]
plt.boxplot(data)
plt.grid()
```

### Additional Information

A **Box Plot**, also known as a **Whisker Plot**

---

##### Interquartile Range (IQR)
`IQR = Q3 - Q1`

---

##### Outlier Boundaries
`Lower = Q1 - (1.5 × IQR)`  
`Upper = Q3 + (1.5 × IQR)`

---

##### Key Points
- Values outside the bounds are **outliers**  
- The box represents the **middle 50% of data**  
- The median shows the **center of distribution**

### 1. Multiple Datasets ::

```python
np.random.seed(0)
group1=np.random.normal(50, 10, 100)
group2=np.random.normal(60, 15, 100)

plt.boxplot([group1, group2], tick_labels=["Group-1", "Group-2"])
plt.title("Comparison of Two Groups Data")
plt.grid(True)
plt.show()
 
```

### 2. Horizontal Box Plot ::

```python
np.random.seed(0)
classA = np.random.normal(70, 8, 120)   # mean=70, std=8
classB = np.random.normal(65, 12, 120)  # mean=65, std=12

plt.boxplot([classA, classB], tick_labels=["Class A", "Class B"],vert=False)
plt.title("Comparison of Students' Marks")
plt.grid(True)
plt.show()
```

### 3. Show Mean ::

```python
np.random.seed(0)
group1=np.random.normal(50, 10, 100)
group2=np.random.normal(60, 15, 100)

plt.boxplot([group1, group2], tick_labels=["Group-1", "Group-2"],showmeans=True)
plt.title("Comparison of Two Groups Data")
plt.grid(True)
plt.show()
 
```

### 4. Change Whisker Length ::

```python
group1 = np.random.normal(50, 10, 100)
group2 = np.random.normal(60, 15, 100)

plt.boxplot([group1, group2], tick_labels=["Group-1", "Group-2"], vert=False, showmeans=True, whis=2.0)
plt.title("Grouping Multiple Data.")
plt.grid(True)
plt.show()
```
