# Num_Py Arrays vs Python List
 - Most time taken between Numpy array and Python List
 - Most Memory usage between Numpy array and Python List

```python
import numpy as np
import time
```

```python
arr = np.array([1,2,3,4,5]) #python list convert to np array
print(arr)
print(type(arr))
```

```python
size = 10_000_00

#python list
py_list = list(range(size))
start = time.time()
sq_list=[x**2 for x in py_list] 
end =time.time()
print(f"Python list time = {end-start} seconds")

#numpy arrays
np_array=np.array(py_list)
start = time.time()

#vectorization-> avoiding explicit loops (for, while) and using NumPyâ€™s built-in array operations.
sq_array = np_array ** 2
end=time.time()
print(f"numpy array time = {end-start} seconds")
```

```python
#memory occupies information between python list and Numpy
import sys

print(f"python list size = {sys.getsizeof(py_list) *len(py_list)} bytes")
print(f"Numpy array size = {np_array.nbytes} bytes")
```
