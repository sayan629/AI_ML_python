# Common DataTypes of NumPy
 - Integer
 - Floating Point
 - Boolean
 - Complex
 - String
 - Object

```python
import numpy as np
```

```python
arr = np.array([1,2,3,4,5])
print(arr,arr.dtype)
```

```python
arr1 = np.array([1,2.2,3,4,5])
print(arr1,arr1.dtype)
```

```python
arr2 = np.array([10, 5, 20, 3, 15])
bool_arr = arr2 > 10
print(bool_arr)
print(type(bool_arr))
```

```python
arr3 = np.array([1,2,3,4,5,"Sayan"])
print(arr3,arr3.dtype)
```

```python

arr4=np.array([3+5j])
arr5=np.array([2+3j])
print(arr4,arr4.dtype)
print(arr4 + arr5)
print(arr4 - arr5)
```

```python
arr6= np.array(["prime",{1,2,3},3.14])
print(arr6, arr6.dtype)
```
