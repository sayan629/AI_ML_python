# Operations on NumPy Arrays
 - Reshaping
 - Indexing (1D & 2D array)
 - Fancy & Boolean Indexing
 - Slicing
 - pylist COPY vs nparray VIEW

```python
import numpy as np
```

```python
#Reshaping array

arr = np.array([[1,2,3],[4,5,6]])
print(arr, arr.shape ,"\n")

# Reshape the array from (2×3) to (3×2)
reshape_arr=arr.reshape((3,2))
print(reshape_arr, reshape_arr.shape,"\n")


# Flatten the 2D array into a 1D array
flattend=arr.flatten() #2D => 1D
print(flattend,flattend.shape)
```

```python
#Indexing of an Array

print("Indexing of 1D Array")
arr=np.array([1,2,3,4,5])
print(arr[0])
print(arr[3])

print("\n")

print("Indexing of 2D Array")
arr1 = np.array([[1,2,3],[4,5,6]])
print(arr1[0][1])
print(arr1[1][2])
```

```python
#Fancy Indexing=>select multiple elements at once using arrays or lists of indices, instead of simple slices.
arr2 = np.array([1,2,3,4,5])
idx = [0,1,4]
print(arr2[idx])
```

```python
#Boolean Indexing=> select elements based on conditions using a Boolean array
arr3 = np.array([1,2,3,4,5,6,7])
print(arr3[arr3 > 2])
print(arr3[arr3 %2==0])
print(arr3[arr3 %2!=0])
```

```python
#slicing
arr4 = np.array([1,2,3,4,5,6,7])

print(arr4 [1:4]) #[start,end]
print(arr4 [:4]) #[ : end]
print(arr4 [3:]) #[start :]
print(arr4 [0:7:2]) #[start,end,step]
```

```python
#copy vs View
    #In pylist array creates an copy of array
        #In np array, it provides the view[original value]

nums=[1,2,3,4,5]
sub_list = nums[1:3]
print(sub_list)
sub_list[0]=200
print(sub_list) 

print(nums,"\n")

arr5=np.array([1,2,3,4,5])
sub_arr5=arr5[1:3]
print(sub_arr5)
sub_arr5[0]=200
print(sub_arr5)
print(arr5)
```
