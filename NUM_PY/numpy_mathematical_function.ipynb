### Important Mathematical Function
 ##### Aggregation (Sum,Prod,min,max,mean.std,var,meadian,argmin,argmax)
 ##### Power  (square, sqrt, pow(a,b))
 ##### Rounding (round, ceil, floor, trunc)
 ##### Log & Exponential (log, log10, log2, exp)
 ##### Some Other Function (unique, sort, abs)

```python
import numpy as np
```

```python
#mathematical fnx => #aggregration

arr=np.array([1,2,3,4,5])

print(np.sum(arr))      # Sum of all elements => 1+2+3+4+5
print(np.prod(arr))     # Product of all elements => 1*2*3*4*5
print(np.min(arr))      # Minimum value in the array
print(np.max(arr))      # Maximum value in the array
print(np.mean(arr))     # Mean (average) of the elements
print(np.std(arr))      # Standard deviation (spread of data)
print(np.var(arr))      # Variance (square of standard deviation)
print(np.median(arr))   # Median (middle value after sorting)
print(np.argmin(arr))   # Index of the minimum value
print(np.argmax(arr))   # Index of the maximum value
```

```python
#power function
arr1=np.array([1,2,3,4,5])

print(np.square(arr1))   # Squares each element => x^2
print(np.sqrt(arr1))     # Square root of each element => âˆšx
print(np.pow(arr1, 3))   # Raises each element to the power of 3 => x^3
```

```python
#Log and Exponential Function
arr2=np.array([1,2,3,4,5])

print(np.log(arr2))     # Natural logarithm (base e) of each element
print(np.log10(arr2))   # Logarithm base 10 of each element
print(np.log2(arr2))    # Logarithm base 2 of each element
print(np.exp(arr2))     # Exponential function => e^x for each element
```

```python
#Rounding Function => These functions are used to control decimal values.

print(np.round(3.14))   # Rounds to nearest integer (default 0 decimal places)
print(np.ceil(3.14))    # Rounds UP to the next integer (ceiling value)
print(np.floor(3.14))   # Rounds DOWN to the previous integer (floor value)
print(np.trunc(3.14))   # Removes decimal part (truncates toward zero)
```

```python
#Some Extra Function
arr3=np.array([1,2,8,3,4,5,2,3,4])
print(np.unique(arr3))   # Returns unique (distinct) elements in sorted order
print(np.sort(arr3))     # Returns the array sorted in ascending order

arr4=np.array([1,2,-8,3,-4,5,2,-3,4,-9])

print(np.abs(arr4))      ## Returns absolute values (converts negatives to positive)
```
