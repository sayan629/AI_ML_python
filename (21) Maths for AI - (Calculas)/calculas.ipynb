```python
import numpy as np
import matplotlib.pyplot as plt
```

### Plot f(x)= √x

```python
#Define x values ( x>= 0 for sqrt)
x= np.linspace(0, 100, 100_000)
y= np.sqrt(x)

#plot
plt.plot(x, y, label= "y =√x")
plt.xlabel("x")
plt.ylabel("y")
plt.title("y= √x")
plt.legend()
plt.grid(True)
plt.show()
```
