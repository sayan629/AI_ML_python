# Single Neuron Using PyTorch

```python
import torch
import torch.nn as nn
```

### Step 1: Create Input 

```python
torch.manual_seed(42)  #generate same random number after each run
inputs = torch.tensor([1.0, 2.0, 3.0])
```

### Step 2: Create a Single Neuron

Create one linear neuron with:
- Input Features = 3
- Output Features = 1

```python
neuron = nn.Linear(in_features=3, out_features=1)
```

### Step 3: Pass the input through the neuron to calculate the output.

```python
output = neuron(inputs)
```

```python
output
```

### Step 4: View the Weights

```python
neuron.weight
```

### Step 5: View the Bias

```python
neuron.bias
```

### Step 6: Verify the Linear Equation
 - y=Wx+b

```python
# verify using .. Formula
neuron.weight @ inputs + neuron.bias
```
