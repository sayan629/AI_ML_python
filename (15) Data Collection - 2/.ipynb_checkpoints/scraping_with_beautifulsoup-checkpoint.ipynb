# 🌐 Web Scraping using Python (Requests + BeautifulSoup)

## 📌 Objective
Extract country names and population from a website and store them in a CSV file.

---

## 🧰 Libraries Used
- requests → to fetch webpage  
- BeautifulSoup → to parse HTML  
- pandas → to store data  

---

## 🚀 Step 1: Fetch Website Data
- Send a request to the target website  
- Check if the response status code is 200 (success)  
- Access:
  - HTML content  
  - Raw data  
  - Headers (metadata)

---

## 💾 Step 2: Save HTML File
- Store the fetched HTML content into a local file  
- Helps in offline parsing and debugging  

---

## 🔍 Step 3: Parse HTML
- Load the saved HTML file  
- Create a BeautifulSoup object  
- Use a parser like lxml  
- Extract basic elements like heading tags  

---

## 📊 Step 4: Extract Data
- Find all relevant tags (e.g., h3 for country names)  
- Loop through each element  
- Extract:
  - Country name  
  - Population from related tags  
- Store data in a list  

---

## 📁 Step 5: Create Data Structure
- Convert extracted list into a tabular format  
- Use a DataFrame structure  
- Define columns like:
  - Name  
  - Population  

---

## 📤 Step 6: Save Data
- Export the cleaned data into a CSV file  
- Disable index for clean output  

---

## ⚠️ Common Mistakes
- Using print while assigning values (returns None)  
- Incorrect tag selection  
- Missing encoding while reading/writing files  

---

## 🔄 Workflow Summary
1. Fetch website  
2. Save HTML  
3. Parse HTML  
4. Extract data  
5. Store in structure  
6. Export CSV  

---

## 🎯 Outcome
- Extracted dataset of countries and populations  
- Stored in a structured CSV file  

---

## ✅ Conclusion
This project demonstrates:
- Basic web scraping  
- HTML parsing  
- Data extraction  
- Data storage and export

```python
# Web Scrapping

import requests   
URL = "https://www.scrapethissite.com/pages/simple/"  # URL of the website to scrape
res=requests.get(URL)  # Send GET request to the URL
```

```python
# Check if request was successful (status code 200 = OK)
if res.status_code == 200:
    # print(res.text)  # HTML content of the page as a string
    # print(res.content) # Raw bytes of the response
    print(res.headers)  # Meta data like content type,server info etc.


# Save the HTML content into a file
with open("scraped_data/data1.html","w",encoding="utf-8") as f:
    f.write(res.text)
```

```python
# BeautifulSoup

from bs4 import BeautifulSoup

# Open the saved HTML file
with open("scraped_data/data1.html","r") as f:
    html_content=f.read()

# Create BeautifulSoup object
soup = BeautifulSoup(html_content,"lxml")
```

```python
soup.find("h1") # Find first <h1> tag
```

```python
all_h3=soup.find_all("h3") # Find all <h3> tags (country names)

all_countries = []
for h3 in all_h3:
   name= h3.get_text(strip=True)  # Extract country name (remove extra spaces)
   
    # print(h3.find_parent("div")["class"])

    
    # Find the next <div> and extract population
   population=print(h3.find_next("div").select("span.country-population")[0].get_text(strip=True))
   print(h3.find_next("div").select_one("span.country-population").get_text(strip=True))

   print(h3.attrs) # Print attributes of h3 tag

   # Append data to list
   all_countries.append([name,population])
```

```python
import pandas as pd
# Create DataFrame
df=pd.DataFrame(all_countries,columns=["Name","Population"])
```

```python
df.to_csv("cleaned_data/data.csv",index=False)
```
