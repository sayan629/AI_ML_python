# 📚 Stephen King Books API → Pandas DataFrame

This demonstrates how to fetch data from a public API and convert it into a structured Pandas DataFrame.
---

## 🚀 Requirements
- Python 
- `requests` library
- `pandas` library

Install dependencies:

```bash
pip install requests


````python
import requests
````

````python
URL = "https://stephen-king-api.onrender.com/api/books" #  API URL
res=requests.get(URL)  #Fetch data
````

````python
json_data=res.json() #convert to json
````

````python
import pandas as pd
df=pd.json_normalize(json_data["data"]) #Normalize JSON to DataFrame
df=df[['id','Title','Year','ISBN']]
````
