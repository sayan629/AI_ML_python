### Grouping and Aggregation in Pandas

**Grouping and Aggregation** are used to analyze data by grouping rows based on a column and applying summary functions.

**`groupby()`** → Groups the DataFrame based on one or more columns.

**Common aggregation functions:**
- `mean()` → Average value  
- `max()` → Maximum value  
- `min()` → Minimum value  
- `sum()` → Total value  
- `count()` → Number of records     etc.... 

**`agg()`** → Used to apply multiple aggregation functions on a particular column or different functions on different columns.

**`aggregate()`** →  alias of ***`agg()`*** , same task perform

```python
import pandas as pd
```

```python
df=pd.read_csv("raw_data.csv")
```

```python
#.........Grouping and Aggregation...........

df.groupby("country") ["income"].mean()
df.groupby("country") ["income"].min()
df.groupby("country") ["income"].max()
df.groupby("country") ["income"].sum()
df.groupby("country") ["income"].count()

df.groupby("gender") ["income"].mean()
df.groupby("gender") ["income"].max()

#Apply multiple aggregation functions 
df.groupby("country") ["income"].agg(["mean","min","max"])
df.groupby("country") ["income"].aggregate(["mean","min","max"]) # aggregate() is alias of agg()

df.groupby("country") ["income"].aggregate(avg_salary= "mean",min_salary= "min",max_salary="max")

#perform diffrent aggregate function on different columns
df.groupby("country").agg({
    "income":"max",
    "age": "mean"
})

df.groupby("country").agg(
    max_salary=("income","max"),
    avg_age=("age","mean")
)
```
