### Data Concatenation

Data concatenation is used to combine multiple DataFrames either **row-wise** or **column-wise**.

* **Row-wise Concatenation**
  Combines rows from multiple DataFrames into a single DataFrame. By default, the original indexes are preserved.

* **Ignore Index**
  Resets the index after concatenation and creates a new sequential index.

* **Column-wise Concatenation**
  Combines DataFrames side by side by adding columns. Each DataFrame is aligned based on their index.


```python
import pandas as pd
```

```python
#Data Concatenation

df1=pd.DataFrame({
    "id":[1,2,3],
    "name":["Sayan","Samata","Sreenika"]
})

df2=pd.DataFrame({
    "id":[4,5,6],
    "name":["Lakshmi Narayan","Manju","Aayansh"]
})

pd.concat([df1,df2]) #row wise concatenation
pd.concat([df1,df2],ignore_index=True) # ignore_index=True creates a new sequential index

pd.concat([df1,df2],axis=1) #column wise concatenation
```
