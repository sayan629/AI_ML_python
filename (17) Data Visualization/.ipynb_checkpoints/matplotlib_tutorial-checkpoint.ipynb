## Data Visualization
 #### Plot a Data(using Matplotlib) :

```python
import matplotlib.pyplot as plt
```

```python
# ---------------- BASIC LINE PLOT ----------------
x=[1,2,3,4]   # X-axis values
y=[5,6,7,8]   # Y-axis values

plt.plot(x,y) #Line Plot  
plt.show()   # Display graph
```

```python
#Movies Data
# Oscar Winner Movies data
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.plot(years, oscar_revenue)
plt.title("Oscer Movies Revinue in Each Year") # Title
plt.xlabel("Years")            # X-axis label
plt.ylabel("Revenue (in $M)")  # Y-axis label
```

```python
#Plot Multiple
# Non-Oscar Winner Movies

non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

# Compare both datasets
plt.plot(years, oscar_revenue, label="Oscar Movies")
plt.plot(years, non_oscar_revenue, label="Non-Oscar Movies")

plt.title("Oscar Movies vs Non-Oscar Movies Revenues:")
plt.xlabel("Years")
plt.ylabel("Revenue (In $M)")
plt.legend() # legend(): Help the viewer understand which line represents what data
# plt.legend(loc="upper right")
# plt.legend(loc="upper left")
# plt.legend(loc="center")
```

```python
# Non-Oscar Winner Movies

non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

#using format Strings
plt.plot(years, oscar_revenue, "o--r", label="Oscar Movies")
plt.plot(years, non_oscar_revenue, "<-y", label="Non-Oscar Movies")

plt.title("Oscar Movies vs Non-Oscar Movies Revinues:")
plt.xlabel("Years")
plt.ylabel("Revenue (In $M)")
plt.legend(loc="upper right")
```

```python
non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

# ---------------- ADVANCED CUSTOMIZATION ----------------
plt.plot(years, oscar_revenue, color="red", marker="<",linestyle="-.",linewidth=3, label="Oscar Movies")
plt.plot(years, non_oscar_revenue, label="Non-Oscar Movies")

plt.title("Oscar Movies vs Non-Oscar Movies Revenues:")
plt.xlabel("Years")
plt.ylabel("Revinue (In $M)")
plt.legend(loc="upper right")
```

```python
plt.plot(years, oscar_revenue, label="Oscar Movies")
plt.plot(years, non_oscar_revenue, label="Non-Oscar Movies") 
plt.title("Oscar Movies vs Non-Oscar Movies Revenues:")
plt.xlabel("Years")
plt.ylabel("Revinue (In $M)")
plt.legend()

# ---------------- STYLE & GRID ----------------
print(plt.style.available)
# plt.style.use("dark_background")  #Show available styles
plt.style.use("default") # Apply style
plt.grid()  # Add grid
plt.show()
```

### Bar Charts

```python
movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]

plt.bar(years,oscar_revenue,color="green")
plt.title("Revenue is each Year for Oscar Movies:")
plt.xlabel("Year")
plt.ylabel("Revenue (in $M)")

for i in range(len(years)):
    plt.text(years[i],oscar_revenue[i]+20,str(oscar_revenue[i]),ha="center")

plt.ylim(0,max(oscar_revenue)+200) # Set Y-axis limit
```

```python

oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_revenue = [1005, 170, 427, 133, 232]

non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]

years = [2008, 2009, 2010, 2011, 2012]

#plotting Bar char
import numpy as np
x=np.arange(len(years)) # Create positions
width=0.4  # Width of bars

plt.bar(x-width/2,oscar_revenue,width,label="Oscar Movies") #SHIFTING to LEFT
plt.bar(x+width/2,non_oscar_revenue,width,label="Non-Oscar-Movies") #SHIFTING to RIGHT

plt.title("Oscar Vs Non-Oscar Movies Revenue")
plt.xlabel("Year")
plt.ylabel("Revenue ( in $M):")
plt.legend()
plt.xticks(x,years)  # Set year labels
```

```python
# ---------------- HORIZONTAL BAR ----------------
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_revenue = [1005, 170, 427, 133, 232]
plt.barh(oscar_movies,oscar_revenue)
plt.xlabel("Revenue")
plt.ylabel("Movie Names")
```

 ### Scatter Plots

```python
#Scatter Plot

people=["Person A","Person B","Person C","Person D","Person E",
        "Person F","Person G","Person H","Person I","Person J"]

age=[22,25,30,35,40,45,50,55,60,65]
blood_pressure=[110,115,120,122,125,130,135,123,145,150]

plt.scatter(age,blood_pressure)
plt.title("Age vs Blood Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")

plt.grid()
```

```python
# Cutomization of Scatter plots
people=["Person A","Person B","Person C","Person D","Person E",
        "Person F","Person G","Person H","Person I","Person J"]

age=[22,25,30,35,40,45,50,55,60,65]
blood_pressure=[110,115,120,122,125,130,135,123,145,150]

# colors = ["Red", "Blue", "Green", "Yellow", "Orange", "Purple", "Pink", "Black", "White", "Brown"]

colors=["green" if x<135 else "red" for x in blood_pressure]

# plt.scatter(age,blood_pressure,color=colors,sizes=blood_pressure,marker="<",alpha=0.5)
plt.scatter(age,blood_pressure,sizes=blood_pressure,cmap="OrRd",c=blood_pressure) 
plt.title("Age vs Blood Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()
plt.colorbar(label="BP")

#anotations
for i in range(len(people)):
    plt.annotate(people[i], xy=(age[i],blood_pressure[i]), xytext=(age[i]+1,blood_pressure[i]+1))

plt.xlim(min(age),max(age)+10)
plt.ylim(min(blood_pressure),max(blood_pressure)+5)
```

```python
#Multiple Plots
cities = ["Bhubaneswar", "Delhi", "Mumbai", "Kolkata", "Chennai"]

#Winter: Temparature(c) vs Humidity(%)
winter_temp=[5,2,10,0,7]
winter_humidity=[60,50,55,45,65]

#Summer: Temparature(c) vs Humidity(%)
summer_temp=[25,30,28,35,27]
summer_humidity=[60,50,55,45,65]

plt.scatter(winter_temp,winter_humidity, label="Winter")
plt.scatter(summer_temp,summer_humidity, label="Summer")
plt.title("Summer vs Winter")
plt.xlabel("Temparature (C)")
plt.ylabel("Humidity(%)")
plt.legend()
```

```python
#Pie Chart
expenses=["Salaries","Rent","Marketing","R&D","Miscellaneous"]
amounts=[500,150,120,100,50]

plt.pie(amounts,labels=expenses)
```

```python
#Customizations on Pie Chart
expenses=["Salaries","Rent","Marketing","R&D","Miscellaneous"]
amounts=[500,150,120,100,50]


# colors=["red","black","pink","blue","yellow"]
# plt.pie(amounts,labels=expenses,autopct="%1.1f%%",colors=colors)
plt.pie(amounts,labels=expenses,autopct="%1.1f%%",
       wedgeprops={
           "edgecolor":"black",
           "linewidth": 1,
           "linestyle":"--"
           # "alpha":0.5,
           # "fill":False
           
       })
plt.show()
```

```python
#Customizations on Pie Chart
expenses=["Salaries","Rent","Marketing","R&D","Miscellaneous"]
amounts=[500,150,120,100,50]

explode=[0,.1,0,0.2,0]

plt.pie(amounts,labels=expenses,autopct="%1.1f%%",
       wedgeprops={
           "edgecolor":"black",
           "linewidth": 1,
           "linestyle":"--"
       },
       explode=explode,
        startangle=45, # startangle used to rotate the chart
       shadow= True
       )
plt.show()
```
