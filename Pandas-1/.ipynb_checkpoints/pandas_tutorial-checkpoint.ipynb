## 📊 Pandas DataFrame Creation Example

### 🔹 Objective
    Create a simple DataFrame using a Python dictionary and display the data.

### 🔹 Steps Performed
1. Imported the pandas library.
2. Created a dictionary named `info` with:
   - Name
   - CGPA
3. Converted the dictionary into a Pandas DataFrame.
4. Printed the DataFrame.
5. Displayed the DataFrame directly in the notebook.


```python
import pandas as pd
```

```python
# Creating a dictionary containing student data
info={
    "Name":["Sayan","Shlok","Ruhon"],
    "CGPA":[9.38,7.47,8.95]
}

# Converting the dictionary into a Pandas DataFrame
df=pd.DataFrame(info)
print(df)
```
