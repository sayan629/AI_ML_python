## 📊 Data Frame Creation
A Pandas DataFrame can be created using different data structures:

- **Dictionary of Lists**
  - Keys represent column names.
  - Values represent column data.
  
- **List of Lists**
  - Each inner list represents a row.
  - Column names can be assigned separately.

- **NumPy Array**
  - A NumPy array can be converted into a DataFrame.
  - Column names can be specified using the `columns` parameter.

```python
import pandas as pd
```

```python
#Data Frame Created using "Dictionary of lists"
info={
    "Name":["Sayan","Samata","Shreyas"],
    "Age":[23,22,15],
    "CGPA":[9.5,8.9,7.2]
}
df=pd.DataFrame(info)
print(df)
print()

#print label of rows
print(df.index)

#print label of columns
print(df.columns)
 
```

```python
#Data Frame Created using "List of lists"
df=pd.DataFrame([["Adam",23],["Bob",25],["Eve",21]],columns=["Name","Age"])
print(df)
```

```python
#Data Frame Created using "numpy array"

import numpy as np

# Creating a 2D NumPy array
arr=np.array([[1,2,3],[4,5,6]])  

# 'columns' parameter is used to assign column names
df=pd.DataFrame(arr,columns=["A","B","C"])
print(df)
```
