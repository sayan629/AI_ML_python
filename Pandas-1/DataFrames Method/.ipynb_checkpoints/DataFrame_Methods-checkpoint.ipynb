## 📊 Pandas DataFrame Basic Operations

This program demonstrates how to **read datasets and explore them using Pandas DataFrame methods**.

### 📂 Reading Data
- `pd.read_csv()` → Reads data from a **CSV (Comma Separated Values) file**.
- `pd.read_json()` → Reads data from a **JSON (JavaScript Object Notation) file**.

### 🔍 Data Preview
- `head()` → Shows the **first rows** of the dataset.
- `tail()` → Shows the **last rows** of the dataset.
- `sample()` → Shows **random rows** from the dataset.

### 🧾 Dataset Information
- `info()` → Displays **structure of the dataset** (columns, data types, non-null values).
- `shape` → Returns **(rows, columns)** of the DataFrame.
- `columns` → Lists all **column names**.

### 📈 Statistical Summary
- `describe()` → Shows important statistics of numerical columns  
  *(count, mean, std, min, 25%, 50%, 75%, max)*.
- `nunique()` → Counts **unique values** in each column.

### ⚠️ Missing Data Handling
- `isnull()` → Detects **missing values**.
- `isnull().sum()` → Counts total **missing values per column**.

### 📊 Data Analysis
- `value_counts()` → Counts **frequency of values in a column**.
- `sort_values()` → Sorts data **ascending or descending**.

### 🎯 Data Selection
- `loc[]` → Select rows using **labels or conditions**.
- `iloc[]` → Select rows using **index positions**.

These methods help in **quickly understanding and analyzing datasets using Pandas**.

```python
import pandas as pd
import numpy as np
```

```python
# Read CSV file and store it in DataFrame
df=pd.read_csv("employee_data.csv")
print(df)
print()

#Read JSON file and store it in DataFrame
npp=pd.read_json("employee_data.json")
print(npp)
```

```python
# Methods in DataFrame

# Print first 5 rows (default)
print(df.head(), "\n")

# Print first 3 rows
print(df.head(3), "\n")

# Print last 5 rows (default)
print(df.tail(), "\n")

# Print last 2 rows
print(df.tail(2), "\n")

# Print 1 random row (different every time you run)
print(df.sample(), "\n")

# Print 5 random rows
print(df.sample(5), "\n")

# Print summary of DataFrame (rows, columns, data types, non-null values)
print(df.info(), "\n")

# Print shape of DataFrame (rows, columns)
print(df.shape, "\n")

# Print statistical summary of numerical columns
# (count, mean, std, min, 25%, 50%, 75%, max)
print(df.describe(), "\n")

# Print all column names
print(df.columns, "\n")

# Print number of unique values in each column
print(df.nunique(), "\n")

# Show missing values in the dataset
print(df.isnull(), "\n")

# Count total missing values in each column
print(df.isnull().sum(), "\n")

# Count frequency of values in Department column
print(df["Department"].value_counts(), "\n")

# Sort data by Salary (ascending)
print(df.sort_values("Salary"), "\n")

# Sort data by Salary (descending)
print(df.sort_values("Salary", ascending=False), "\n")

# Select rows where Age > 30
print(df.loc[df["Age"] > 30], "\n")

# Select first 3 rows using index position
print(df.iloc[0:3])
```
