## 📊 Pandas Data Selection & Filtering

### Dataset Overview
The dataset `globalAirQuality.csv` is loaded into a Pandas DataFrame.  

### 📌 Selecting Columns
- Single column → returns a **Series**
- Multiple columns → returns a **DataFrame**

### 📍 Selecting Rows
**loc (Label Based)**
- Uses row labels
- Slice is **inclusive** (end index included)

**iloc (Index Based)**
- Uses integer positions
- Slice is **exclusive** (end index excluded)

### 🔎 Accessing Values
- `loc` → label-based selection  
- `iloc` → index-based selection  
- `at` / `iat` → faster access for a **single value**

### 🧠 Copy vs View
Using `.copy()` ensures the new variable is **independent** and changes will not affect the original DataFrame.

### 🔍 Filtering Data
Rows can be filtered using conditions:
- Single condition
- Multiple conditions using `&` or `|`

Filtering can also be **chained with column selection**.

### ⚠ Index After Filtering
Filtering keeps the **original index values**, so index `0` may not exist in the filtered result.

### 🔎 Query Method
`query()` allows filtering using **string expressions** similar to SQL.

Key points:
- Use **backticks** for column names with spaces
- Use **@** to reference Python variables
- `query()` **always returns a copy**

```python
  import pandas as pd
```

```python
df=pd.read_csv("globalAirQuality.csv")

df.describe()
```

```python
#selecting the data

# Selecting a single column
df["city"]  # Returns a Series (VIEW of original data, not a copy)

# Selecting multiple columns
df[["city","aqi"]]

# SELECTING ROWS (Label Based vs Index Based)

df.loc[0]  # loc → Label based selection
df.loc[0:2] #start idx:end idx(inclusive)

df.iloc[2] # iloc → Integer position based selection
df.iloc[0:2] #start idx:end idx(end index excluded)

#individual cells - rows,column

df.loc[0,"aqi"] 
df.loc[0,["city","aqi"]]
df.loc[0:2,["city","aqi"]]

# NOTE:
# iloc does NOT accept column names (strings)
# It only works with integer positions
# df.iloc[0,"city"]  ❌ invalid

df.columns
df.iloc[0:3,2]  # Select rows 0–2 and column at position 2

df.loc[0:2,["country","city","latitude","longitude"]]  # Select specific columns using labels
df.iloc[0:3,1:5    # Select using index positions

#select single Scalar value - at & iat
df.at[0,"city"] # at → Fast label-based access for a single value
df.iat[0,2]     # iat → Fast integer-position-based access

# VIEW vs COPY CONCEPT

# Create a copy of the "city" column
# Using .copy() ensures modifications don't affect original DataFrame
cities=df["city"].copy()

# Modify value in copied Series
cities[0]="Iran"
cities   # This change affects only 'cities', not original df
```

```python
#Filtering of Data
df[df["aqi"]>100] #single condition
df[(df["aqi"]>100) & (df["temperature"]>30)] #multiple condition
df[(df["aqi"]>100) & (df["temperature"]<30)]

#chaining
df[df["aqi"]>100][["city","aqi"]]
df[(df["aqi"]>100) & (df["temperature"]>30)][["city","aqi"]]

aqi_data=df[(df["aqi"]>100) & (df["temperature"]>30)][["city","aqi"]]
aqi_data.iloc[0]  # Access first row using index position


# NOTE:
# aqi_data.loc[0] may fail because
# after filtering the original index values remain
# Example index could be: 3,6,9 instead of 0,1,2

aqi_data.loc[6]
```

```python
# Query rules

# Expression written as a string with column reference
df.query("aqi>100") 

# If column names contain spaces or special characters
# use backticks (`column name`)
df.query("`temperature` > 30")

#operators
df.query("aqi>100 & temperature>30")

#chaining with query
df.query("aqi>100 & temperature>30")[["city","aqi"]]

# @ to reference Python Vars
aqi_vals=100
df.query("aqi>@aqi_vals & temperature>30")[["city","aqi"]]

#Remember: Queryreturns a COPY of the data (not a view)
aqi_data=df.query("aqi>@aqi_vals & temperature>30")[["city","aqi"]].copy()
aqi_data

```
