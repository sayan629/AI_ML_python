### Series in Pandas

### Properties of Series

```python
import pandas as pd
```

```python
s=pd.Series([1,2,3,4,5])
print(s)
print(type(s))

# ------------ Accessing Using Index Position -------------
print(s[0])
print(s[2])
```

```python
#------------- Creating Series with Custom Labels --------------
s=pd.Series([22,23,24,25], index=["Carter","Shield","Blake","Grant"])
print(s)

# Accessing data using label
print(s["Carter"])

# Printing all label names
print(s.index)
```

### 📌 Properties in Series
- **Homogeneous Data (Generally)**  
  - Mostly stores the same data type.  
  - Can also store mixed data types (object type).

- **Handle Missing Values with NaN (None)**  
- **Mutable Values**  
  - Values inside a Series can be modified after creation.

- **Immutable Size**  
  - Length of a Series cannot be changed directly.  
  - `drop()` creates a new Series instead of modifying the original one.

- **Vectorized Operations**  
  - Supports element-wise operations like addition, subtraction, multiplication, etc.

```python
# --------------Homogeneous / Heterogeneous --------------
# Series can hold mixed data types (not strictly homogeneous)

s=pd.Series([22,23,24,25,"abc"], index=["Carter","Shield","Blake","Grant",25])
print(s)
```

```python
#Vectorized Operations
s1=pd.Series([22,24,25,10])
s2=pd.Series([20,23,26,90])
print(s1+s2)  # Element-wise addition
```

```python
#mutable values but immutable sizes
s1=pd.Series([22,24,25,10])
s1[2]=30
print(s1)
```

```python
#drop
s1=pd.Series([22,24,25,10])

s1[0]=100  #Modify value

changed_s1=s1.drop(0) # Drop index 0 (creates new Series)

print(s1)  # Original Series
print(changed_s1) # After dropping index 0
```
