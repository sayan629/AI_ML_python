```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("shop_smart_ecommerce.csv")
```

```python
df.head()
df.describe()
df.isnull().sum()
df.info()
```

```python
target = df['Revenue'].value_counts()
sns.countplot(
    x='Revenue',
    data = df
)
plt.show()
```

### Feature Processing

```python
# Convert Boolean Columns in int type

df["Weekend"] = df["Weekend"].astype(int)
df["Revenue"] = df["Revenue"].astype(int)
```

```python
df.info()
```

```python
# Encode Categorical Variables

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df["VisitorType"] = le.fit_transform(df["VisitorType"])
df["Month"] = le.fit_transform(df["Month"])
```

```python
df.info()
```

### Train-Test-Split

```python
X= df.drop('Revenue',axis=1)
y = df['Revenue']
```

```python
X_train,X_test, y_train,y_test = train_test_split(
    X,y,test_size=0.3,random_state=42, stratify=y
)

# stratify=y keeps class proportions balanced.
```

```python
X_train.head()
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
# Evaluation
from sklearn.metrics import accuracy_score,f1_score

y_pred = model.predict(X_test) #Prediction

print("Accuracy Score:" ,accuracy_score(y_test,y_pred))
print("f1 Score:" ,f1_score(y_test,y_pred))
```

```python
# Plot Tree

from sklearn.tree import plot_tree
plt.figure(figsize=(18,10))
plot_tree(
    model,
    feature_names = X.columns,
    class_names = ['No Purchase', 'Purchase'], # 0 =>No Purchase, 1 =>Purchase
    filled = True,
    max_depth=2
)
plt.tight_layout()
plt.show()
```

### Decision Tree with Post-prunning

```python
full_tree = DecisionTreeClassifier(random_state=42)
full_tree.fit(X_train,y_train)
```

```python
# Find Best Alpha
path = full_tree.cost_complexity_pruning_path(X_train,y_train)
ccp_alphas = path.ccp_alphas
print(ccp_alphas)
```

```python
# Train our model for all alphas
trees = []
for alpha in ccp_alphas:
    model = DecisionTreeClassifier(random_state=42, ccp_alpha = alpha)
    model.fit(X_train,y_train)
    f1 = f1_score(y_test, y_pred)
    trees.append((model,alpha,f1))
```

```python
best_f1 =0
best_alpha =0

for model, alpha,f1 in trees:   
    if f1>best_f1:
        best_f1 = f1
        best_alpha = alpha

print("Best Alpha:", best_alpha)
print("Best F1 Score:", best_f1)
```

```python
best_model = DecisionTreeClassifier(ccp_alpha=best_alpha)
best_model.fit(X_train,y_train)
```

```python
plt.figure(figsize=(18,10))
plot_tree(
    best_model,
    feature_names = X.columns,
    class_names = ['No Purchase', 'Purchase'],
    filled = True
)
plt.title(f"Pruned Decision Tree (ccp_alpha={best_alpha:.6f})")
plt.tight_layout()
plt.show()
```

```python
print(best_model.score(X_test,y_test))
```
