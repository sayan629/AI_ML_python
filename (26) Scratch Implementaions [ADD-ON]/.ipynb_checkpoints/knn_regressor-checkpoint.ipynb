## KNN Regressor From Scratch

```python
import numpy as np
```

```python
class KNNRegressor:
    def __init__(self, k=3):
        # step 1 : initialize K value
        self.k = k

    def _euclidean_dis(self,x1,x2):
        # step 2 : calculate Euclidean distance
        return np.sqrt(np.sum((x1-x2)**2))

    def fit(self, X,y):   # store training data
        self.X_train = X
        self.y_train = y

    def _predict_one(self,x):
        
        # calculate distance from all training points
        distances = [self._euclidean_dis(x,x_train) for x_train in self.X_train]

        # step 3 : get indices of nearest K neighbors
        knn_indices = np.argsort(distances)[:self.k]

        # get target values of nearest neighbors
        knn_classes = [self.y_train[i] for i in knn_indices]

        # step 4 : take average of nearest neighbors
        prediction = np.mean(knn_classes)
        
        return prediction

    def predict(self,X):
        y_pred = [self._predict_one(x) for x in X]
        return np.array(y_pred)
        
```

```python
# Training data
X_train = np.array([
    [1, 2],
    [2, 3],
    [3, 3],
    [6, 5],
    [7, 7]
])

# Target values
y_train = np.array([0, 0, 0, 1, 1])

# Test data
X_test = np.array([
    [2, 2],
    [6, 6]
])

model = KNNRegressor()   # Create model
model.fit(X_train , y_train) # Train model

y_pred = model.predict(X_test) # Predict
print(y_pred) 
```
