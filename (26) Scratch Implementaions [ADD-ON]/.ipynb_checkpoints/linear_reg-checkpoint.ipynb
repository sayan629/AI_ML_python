```python
import numpy as np
```

```python
# Linear Regression

class LinearRegression:
    def __init__(self, learning_rate=0.01,n_iter=1000):
        self.bias = None
        self.weights = None
        self.learning_rate = learning_rate
        self.n_iter = n_iter

    def fit(self, X, y): #X_train, y_train
        m,n = X.shape # no of samples, no of features

        # Step-1 Initialize Bias & weights with random values
        self.bias = 0
        self.weights = np.zeros(n) #return array [0,0,...0]

        # Gradient Descent
        for i in range(self.n_iter):
            
            # Step-2 calculate y predictions ( y_pred = bias + X.weights )
            y_pred = self.bias + np.dot(X,self.weights)

            #Step-3 - Calculate Gradients
            db = (1/m) * np.sum (y_pred - y )
            dw  = (1/m) * np.dot( X.T, (y_pred - y )) 

            #Step-4 Apply convergence Theorem - parameters update
            self.bias = self.bias - self.learning_rate * db
            self.weights = self.weights - self.learning_rate * dw

        
    def predict(self, X ):
        y_pred = self.bias + np.dot(X,self.weights)
        return y_pred
```

```python
# X = Input Features → must be 2D
# Shape format: (samples, features)
X= np.array ([[1],[2],[3],[4],[5]])

# y = Output/Target values → usually 1D
# Shape format: (samples,)
y= np.array ([2,4,6,8,10])

model = LinearRegression()
model.fit(X,y)

y_pred = model.predict(X)
print(y_pred)

print("Bias Values: ", model.bias)
print("Weights Values: ", model.weights)
```
